```python
set2 = brewer2mpl.get_map('Spectral', 'Diverging', 5).mpl_colors

data_circ       = np.loadtxt("./circ_norm.txt")
bin_edges       = data_circ[:,0]
D1D3_rr         = data_circ[:,1]
D1D4_rl         = data_circ[:,2]

data_rect       = np.loadtxt("./rect_norm.txt")
bin_edges       = data_rect[:,0]
D1D3_hh         = data_rect[:,1]
D1D4_hv         = data_rect[:,2]

data_diag       = np.loadtxt("./diag_norm.txt")
bin_edges       = data_diag[:,0]
D1D3_dd         = data_diag[:,1]
D1D4_da         = data_diag[:,2]


idx = (bin_edges>-.1)&(bin_edges<6)
x 	= bin_edges[idx]
yrr = D1D3_rr[idx]
yrl = D1D4_rl[idx]

yhh = D1D3_hh[idx]
yhv = D1D4_hv[idx]

ydd = D1D3_dd[idx]
yda = D1D4_da[idx]

f = []
inten = []
errs = []

gate_widths = np.linspace(0.01, 14, 50)

for gate in gate_widths:
  
    def integrate(mi, ma):
        idx = (bin_edges>mi)&(bin_edges<ma)
        x 	= bin_edges[idx]
        yrr = D1D3_rr[idx]
        yrl = D1D4_rl[idx]
        
        yhh = D1D3_hh[idx]
        yhv = D1D4_hv[idx]
        
        ydd = D1D3_dd[idx]
        yda = D1D4_da[idx]
        
        idx_1 = (x<(x.min()+gate))
        
        yrrt = yrr[idx_1]
        yrlt = yrl[idx_1]
    
        yhht = yhh[idx_1]
        yhvt = yhv[idx_1]
    
        yddt = ydd[idx_1]
        ydat = yda[idx_1]
    
        g1a = trapz(yrrt)
        g2a = trapz(yrlt)
        dc = (g1a-g2a)/(g1a+g2a)
    
        g1b = trapz(yddt)
        g2b = trapz(ydat)
        dd = (g1b-g2b)/(g1b+g2b)
    
        g1c = trapz(yhht)
        g2c = trapz(yhvt)
        dr = (g1c-g2c)/(g1c+g2c)
        
        fid = (1-dc+dd+dr)/4
        inte = g1a+g2a+g2b+g1b+g1c+g2c
        return fid, inte
        
    integrates = np.array([integrate(-0.1 + i*12.5, 6 + i*12.5) for i in np.arange(-10, 11)])
    fids = integrates[:,0]
    intes = integrates[:,1]
    fid = fids[10]
    inte = intes[10]
    fids = np.delete(fids, 10, 0)
    err = np.std(fids)
    
    
    inten.append([gate, inte])
    f.append([gate, fid])
    errs.append(err)
    
f = np.array(f)
inten = np.array(inten)
errs = np.array(errs)

gate_width = f[:,0]
fidelities = f[:,1]

ix = inten[:,0]
iy = inten[:,1]
np.savetxt('./fidelity_gatewidth.txt', np.vstack((gate_width, fidelities, iy/iy.max(), errs)).T, fmt="%1.5lf")
```

```python
data = np.loadtxt('./fidelity_gatewidth.txt')
gate_widths = data[:,0]
fidelities = data[:,1]
iy = data[:,2]
err = data[:,3]

plt.figure(figsize=(8,6))
ax1 = plt.subplot(111)

ax1.errorbar(gate_widths, fidelities, err, fmt='bo', color=set2[4])

plt.ylabel('Fidelity', fontsize=26)
plt.xlabel('Gate width (ns)', fontsize=26)

plt.xticks([0, 2, 4, 6, 8, 10, 12, 14], fontsize=16)
plt.ylim([0.45, 0.9])
plt.yticks(fontsize=16)

ax2 = ax1.twinx()
ax2.plot(gate_widths, iy/iy.max(), 'r-', color=set2[0], linewidth=4)
ax2.plot(gate_widths, np.zeros_like(gate_widths)-1, 'bo-', color=set2[4], linewidth=4)

plt.ylim([0, 1])
plt.yticks(fontsize=16)


plt.ylabel('Intensity', fontsize=26)
plt.legend(['Intensity', 'Fidelity'], loc=4, fontsize=20)

w = 0.45
h = 0.45
subpos = [1-w,1-h,w,h]
subax1 = add_subplot_axes(ax2,subpos)
subax1.plot(x, yrr, x, yrl)
subax1.legend(['RR', 'RL'], fontsize=20)
subax1.set_xlabel('$\\tau$ (ns)', fontsize=20)
subax1.set_ylabel('$g^{(2)}(\\tau)$', fontsize=20)
subax1.set_xlim([x.min(), x.max()])



plt.savefig('./fidelity_gatewidth.pdf')
plt.savefig('./fidelity_gatewidth.png')
plt.show()
```

```python
        def add_subplot_axes(ax,rect,axisbg='w'):
    fig = plt.gcf()
    box = ax.get_position()
    width = box.width
    height = box.height
    inax_position  = ax.transAxes.transform(rect[0:2])
    transFigure = fig.transFigure.inverted()
    infig_position = transFigure.transform(inax_position)    
    x = infig_position[0]
    y = infig_position[1]
    width *= rect[2]
    height *= rect[3]  # <= Typo was here
    subax = fig.add_axes([x,y,width,height],axisbg=axisbg)
    x_labelsize = subax.get_xticklabels()[0].get_size()
    y_labelsize = subax.get_yticklabels()[0].get_size()
    x_labelsize *= rect[2]**0.5
    y_labelsize *= rect[3]**0.5
    subax.xaxis.set_tick_params(labelsize=x_labelsize)
    subax.yaxis.set_tick_params(labelsize=y_labelsize)
    return subax
```
