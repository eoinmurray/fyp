```python
%pylab inline
from scipy.integrate import trapz
import matplotlib.pyplot as plt
import numpy as np
import brewer2mpl

from scipy.integrate import trapz
from scipy.optimize import curve_fit

set2 = brewer2mpl.get_map('Spectral', 'Diverging', 5).mpl_colors
```

```python
hbar = 6.582119e-16
```

```python
data_circ       = np.loadtxt("./circ_norm.txt")
bin_edges       = data_circ[:,0]
D1D3_rr         = data_circ[:,1]
D1D4_rl         = data_circ[:,2]

data_rect       = np.loadtxt("./rect_norm.txt")
bin_edges       = data_rect[:,0]
D1D3_hh         = data_rect[:,1]
D1D4_hv         = data_rect[:,2]

data_diag       = np.loadtxt("./diag_norm.txt")
bin_edges       = data_diag[:,0]
D1D3_dd         = data_diag[:,1]
D1D4_da         = data_diag[:,2]
```

```python

gate_size=3

def resizer(x, s):
    if x.size%s == 0:
        x = x.reshape(x.size/s, s)
    else:
        x = x[0:-(x.size%s)].reshape(x.size/s, s)
    
    return x

f = []
errs = []

idx = (bin_edges>-2)&(bin_edges<6)
x 	= resizer(bin_edges[idx], gate_size)


for i in xrange(len(x)):
    def foo(mi, ma):
        idx = (bin_edges>mi)&(bin_edges<ma)
        x 	= resizer(bin_edges[idx], gate_size)
        yrr = resizer(D1D3_rr[idx], gate_size)
        yrl = resizer(D1D4_rl[idx], gate_size)
        
        yhh = resizer(D1D3_hh[idx], gate_size)
        yhv = resizer(D1D4_hv[idx], gate_size)
        
        ydd = resizer(D1D3_dd[idx], gate_size)
        yda = resizer(D1D4_da[idx], gate_size)

        g1 = trapz(yrr[i])
        g2 = trapz(yrl[i])
        dc = (g1-g2)/(g1+g2)
    
        g1 = trapz(ydd[i])
        g2 = trapz(yda[i])
        dd = (g1-g2)/(g1+g2)
    
        g1 = trapz(yhh[i])
        g2 = trapz(yhv[i])
        if g1 + g2 == 0:
            dr=0
        else:
            dr = (g1-g2)/(g1+g2)
    
        return x[i,0], (1+dr+dd-dc)/4

    foos = np.array([foo(-2 + j*12.5, 6 + j*12.5) for j in np.arange(-10, 11)])
    f.append(foos[10])
    foos = np.delete(foos, 10, 0)
    err = np.std(foos[:,1])
    errs.append(err)


gate_width_ns = x[0,-1] - x[0,0]

f = np.array(f)
gate_starts = f[:,0]*1e-9
fidelities = f[:,1]

np.savetxt('./fidelity_gatestart.txt', np.vstack((gate_starts/1e-9, fidelities, np.zeros_like(gate_starts) + gate_width_ns)).T, fmt="%1.5lf")
```

```python

data = np.loadtxt('./fidelity_gatestart.txt')
gate_starts = data[:,0]*1e-9
fidelities = data[:,1]

gate_width_ns = data[:,2][0]

def func(x, a, b, c, d):
    return a*np.sin(b*x+c) + d

popt, pcov = curve_fit(func, gate_starts, fidelities, [0.5, 2e9, 0.5, 0.5])
fit_err = np.sqrt(pcov.diagonal())
t = np.linspace(gate_starts.min(), gate_starts.max(), 1000)
gen = func(t, *popt)

plt.figure(figsize=(8,6))

print popt[1], fit_err[1]

plt.errorbar(gate_starts/1e-9, fidelities, errs, fmt='bo', color=set2[4])
plt.plot(t/1e-9, gen, 'r-', color=set2[0], linewidth=4, label='FSS = %0.2lf $\pm$ %0.2lf $\mu eV$'%(popt[1]*hbar/1e-6, fit_err[1]*hbar/1e-6))

plt.ylabel('Fidelity', fontsize=26)
plt.xlabel('Gate start time (ns)', fontsize=26)

plt.legend(fontsize=16)

plt.yticks([-0.2, 0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=16)
plt.xticks([-2, -1, 0, 1, 2, 3, 4, 5, 6], fontsize=16)

plt.savefig('./fidelity_gatestart.pdf')
plt.savefig('./fidelity_gatestart.png')
```
