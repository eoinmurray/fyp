```python
%pylab inline
from scipy.integrate import trapz
import matplotlib.pyplot as plt
import numpy as np
import brewer2mpl

from scipy.integrate import trapz
from scipy.optimize import curve_fit

set2 = brewer2mpl.get_map('Spectral', 'Diverging', 5).mpl_colors
```

```python
data_circ       = np.loadtxt("./circ_norm.txt")
bin_edges       = data_circ[:,0]
D1D3_rr         = data_circ[:,1]
D1D4_rl         = data_circ[:,2]

data_rect       = np.loadtxt("./rect_norm.txt")
bin_edges       = data_rect[:,0]
D1D3_hh         = data_rect[:,1]
D1D4_hv         = data_rect[:,2]

data_diag       = np.loadtxt("./diag_norm.txt")
bin_edges       = data_diag[:,0]
D1D3_dd         = data_diag[:,1]
D1D4_da         = data_diag[:,2]

def plot_correlations():
    linewidth=1
    
    data_circ 	= np.loadtxt("./circ_norm.txt")
    bin_edges 	= data_circ[:,0]
    D1D3_rr 	= data_circ[:,1]
    D1D4_rl 	= data_circ[:,2]
    
    data_rect 	= np.loadtxt("./rect_norm.txt")
    bin_edges 	= data_rect[:,0]
    D1D3_hh 	= data_rect[:,1]
    D1D4_hv 	= data_rect[:,2]
    
    data_diag 	= np.loadtxt("./diag_norm.txt")
    bin_edges 	= data_diag[:,0]
    D1D3_dd 	= data_diag[:,1]
    D1D4_da 	= data_diag[:,2]
    
    plt.figure(figsize=(8,6))
    
    shift=0
    ax1 = plt.subplot(311)
    
    plt.plot(bin_edges+shift, D1D3_rr, color=set2[4], linewidth=linewidth)
    plt.plot(bin_edges, D1D4_rl, color=set2[0], linewidth=linewidth)

    plt.setp(ax1.get_xticklabels(), visible=False)
    plt.legend(['RR', 'RL'], fontsize=18)
    a = np.array(plt.yticks()[0])
    plt.yticks(a[0:-1].astype(np.int32), fontsize=14)
    
    ax2 = plt.subplot(312, sharex=ax1)
    plt.plot(bin_edges+shift, D1D3_dd, color=set2[4], linewidth=linewidth)
    plt.plot(bin_edges, D1D4_da, color=set2[0], linewidth=linewidth)
    plt.ylabel('$g^{(2)}(\\tau)$', fontsize=26)
    plt.legend(['DD', 'DA'], fontsize=18)
    a = np.array(plt.yticks()[0])
    plt.yticks(a[0:-1].astype(np.int32), fontsize=14)
    
    ax3 = plt.subplot(313, sharex=ax1)
    plt.plot(bin_edges+shift, D1D3_hh, color=set2[4], linewidth=linewidth)
    plt.plot(bin_edges, D1D4_hv, 'r-', color=set2[0], linewidth=linewidth)
    plt.setp( ax2.get_xticklabels(), visible=False)
    plt.xlabel('$\\tau$ (ns)', fontsize=26)
    plt.legend(['HH', 'HV'], fontsize=18)
    plt.subplots_adjust(None, None, None, None, None, 0)
    plt.xlim([-40, 40])
    plt.xticks([-37.5, -25, -12.5, 0, 12.5, 25, 37.5], fontsize=14)
#     plt.xlim([-1, 5])
#     plt.xticks([-1, 0, 1, 2, 3, 4, 5], fontsize=14)
    a = np.array(plt.yticks()[0])
    plt.yticks(a[0:-1].astype(np.int32), fontsize=14)
    
    plt.savefig('./correlations.pdf')
    plt.savefig('./correlations.png')
    plt.show()
   
```

```python
plot_correlations()
```

```python

p = 150+12.5/2

def foo(mi, ma):
    idx = (bin_edges>mi)&(bin_edges<ma)
    x 	= bin_edges[idx]
    yrr = D1D3_rr[idx]
    yrl = D1D4_rl[idx]
    
    yhh = D1D3_hh[idx]
    yhv = D1D4_hv[idx]
    
    ydd = D1D3_dd[idx]
    yda = D1D4_da[idx]

    g1 = trapz(yrr)
    g2 = trapz(yrl)
    dc = (g1-g2)/(g1+g2)

    g1 = trapz(ydd)
    g2 = trapz(yda)
    dd = (g1-g2)/(g1+g2)

    g1 = trapz(yhh)
    g2 = trapz(yhv)
    if g1 + g2 == 0:
        dr=0
    else:
        dr = (g1-g2)/(g1+g2)

    return [dr, dd, dc]
```
